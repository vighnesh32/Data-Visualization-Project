```python
# combining IncGeo_BoroughName and Notional Cost (£)
df_colab1 = df_lfb4[["IncGeo_BoroughName","Notional Cost (£)"]]
df_colab1
```

```python
# Checking for missing values
df_colab1.isnull().sum()
```

```python
# filling missing values
df_colab1['Notional Cost (£)'] = df_colab1['Notional Cost (£)'].replace(np.nan, 0)
```

```python
# Checking for missing values
df_colab1.isnull().sum()
```

```python
# taking percent of the notional cost as per the boriughs
df_percent = df_colab1.groupby(["IncGeo_BoroughName"])\
.agg({"Notional Cost (£)" : "sum"})[["Notional Cost (£)"]]\
.apply(lambda x: 100*x/x.sum())\
.sort_values(by="Notional Cost (£)", ascending=False)
```

```python
# display the dataframe
df_percent
```

```python
# reset index
df_percent1 = df_percent.reset_index()
df_percent1
```

```python
# rename columns
df_percent1.rename(columns={'IncGeo_BoroughName': 'boroughs'},inplace=True)
df_percent1.rename(columns={'Notional Cost (£)': 'notionalpercentcost'},inplace=True)
df_percent1
```

```python
# convert dataframe to csv
df_percent1.to_csv(r'C:\Users\vigne\Downloads\df_notionalpercent.csv',index=False)
```

```python
# combining IncGeo_BoroughName and Notional Cost (£)
df_colab2 = df_lfb4[["IncGeo_BoroughName","PumpCount"]]
df_colab2
```
